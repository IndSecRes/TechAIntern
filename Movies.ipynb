```python
df_credits = pd.read_csv('credits.csv', low_memory=False)
```

```python
df_movies_metadata = pd.read_csv('movies_metadata.csv', low_memory=False)
```

```python
df_credits['id'] = df_credits['id'].astype(str)
```

```python
df_movies_metadata['id'] = df_movies_metadata['id'].astype(str)
```

```python
result = pd.merge(df_credits, df_movies_metadata, on='id')
```

```python
result.to_csv('result.csv', index=False, mode='w', header=True)
```

```python
result
```

```python
C= result['vote_average'].mean()
```

```python
C
```

```python
m = result['vote_count'].quantile(0.9)
```

```python
m
```

```python
q_movies = result.copy().loc[result['vote_count'] >= m]
```

```python
q_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
```

```python
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```
